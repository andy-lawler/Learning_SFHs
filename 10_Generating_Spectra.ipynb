# Generating Spectra

Intrinsic spectra generation is handled by the `spectacle` module (see [here](https://github.com/christopherlovell/spectacle) for details). 

Once the data for Illustris and EAGLE have been downloaded, you can run the `spectra_generate.py` script (printed in full below). This can be run in parallel using multi-threading through the [schwimmbad](https://github.com/adrn/schwimmbad) module.

`python run_spectacle.py --ncores=32`

It is also useful to run a number of other data processing tasks in parallel (such as star formation rate binning). These are also listed in this file, and can be commented out if not required.

```python
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.formatters import HtmlFormatter
import IPython

with open('run_spectacle.py') as f:
    code = f.read()

formatter = HtmlFormatter()
IPython.display.HTML('<style type="text/css">{}</style>{}'.format(
    formatter.get_style_defs('.highlight'),
    highlight(code, PythonLexer(), formatter)))
```
